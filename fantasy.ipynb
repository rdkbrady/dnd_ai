```python
import tensorflow as tf
tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True)
from matplotlib import pyplot as plt
from game_state import create_data_for_ai, batch_data_for_ai
from resnet import resnet_v2, lr_schedule
from keras.optimizers import Adam
```

```python
h = 32
w = 32

data_generator = batch_data_for_ai(h,w, 10)
```

```python
n = 3

depth = n * 9 + 2
model = resnet_v2(input_shape=next(data_generator)[0].shape[1:], depth=depth, num_classes = 9)
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=lr_schedule(0)),
              metrics=['accuracy'])
```

```python
model.fit_generator(data_generator, steps_per_epoch = 100, epochs = 200)
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['loss'])
plt.draw()
```

```python
import numpy as np
from matplotlib import pyplot as plt

game_state, requested_action = next(data_generator)
move_probs = model.predict(game_state)[0].reshape(3,3)
print(np.argmax(model.predict(game_state)[0]))
print(np.argmax(requested_action))
fig, axs = plt.subplots(1, 2)

axs[0].imshow(game_state[0])
axs[1].imshow(move_probs)
```
